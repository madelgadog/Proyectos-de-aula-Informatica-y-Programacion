```python
!pip install chempy

from chempy import balance_stoichiometry

reaccion = {}
while True:
    reactivo = input("Ingresa el nombre de un reactivo (o 'fin' para terminar): ")
    if reactivo == 'fin':
        break
    coeficiente = int(input(f"Ingrese el coeficiente para {reactivo}: "))
    reaccion[reactivo] = coeficiente

productos = {}
while True:
    producto = input("Ingrese el nombre de un producto (o 'fin' para terminar): ")
    if producto == 'fin':
        break
    coeficiente = int(input(f"Ingrese el coeficiente para {producto}: "))
    productos[producto] = coeficiente

reaccion_balanceada = balance_stoichiometry(reaccion, productos)[0]

reactivos = {}
for reactivo in reaccion_balanceada.keys():
    cantidad = float(input(f"Ingrese la cantidad de moles de {reactivo}: "))
    reactivos[reactivo] = cantidad

max_producto_por_reactivo = {}
for reactivo, cantidad in reactivos.items():
    max_producto_por_reactivo[reactivo] = cantidad / reaccion_balanceada[reactivo]

reactivo_limitante = min(max_producto_por_reactivo, key=max_producto_por_reactivo.get)
max_producto = max_producto_por_reactivo[reactivo_limitante]
reactivo_en_exceso = {}

for reactivo, cantidad in reactivos.items():
    if reactivo == reactivo_limitante:
        reactivo_en_exceso[reactivo] = cantidad - max_producto
    else:
        reactivo_en_exceso[reactivo] = 0

print("Reactivo limitante:", reactivo_limitante)
```
