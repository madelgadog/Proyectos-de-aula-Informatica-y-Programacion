```python
import numpy as np
from pylab import*
import matplotlib.pyplot as plt
#---Parámetros---
Tmax=35                   #---Tiempo de transcurrido en segundos
dt=0.001                    #---Incremento en el tiempo segundos
m=0.4                      #---Masa en kilogramos
g=9.8                      #---aceleración en m/s2
r=0.9                        #---coeficiente de restitución entre 0 y 1
#----------Condiciones iniciales--------------
H=14                       #---Altura inicial metros
vo=0.0                       # Velocidad de salida (soltar)
#----variables----
T,Y,V=[],[],[] #--Tiempo, posición, velocidad, ec, ep,em
t,y,v=0.,H,vo

while t<=Tmax:
  T.append(t)
  Y.append(y)
  V.append(v)
  if y<0.0:
    v=-r*v
    y=0.
  y=y+v*dt-0.5*g*dt*dt
  v=v-g*dt
  t=t+dt
#---Gráficos------

plt.plot(T,Y,color="red",label='$y(m)$')
plt.xlabel('$t(s)$',fontsize=18)
plt.ylabel('$y(m)$',fontsize=18)
plt.legend(loc='upper right') #---lower center, lower right, upper left
title(u'Pelota saltarina')
plt.show()
```
