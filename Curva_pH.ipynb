```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_titration_curve(Ca, Va, Cb, Vb):
    # pKa de ácido formico
    pKa = 3.75
    Kw = 1.0e-14
    
    # Cálculo de la constante de disociación ácida
    Ka = 10.0**(-pKa)
    
    # Rango de pH y número de puntos
    pH_ini = 0
    pH_fin = 14
    puntos = 100
    pH_valores = np.linspace(pH_ini, pH_fin, puntos)
    
    Vb_valores = []
    for pH in pH_valores:
        H = 10**(-pH)
        OH = Kw / H
        
        # Cálculo de la fracción de titulación
        alpha_A = Ka / (H + Ka)
        Fi = (alpha_A - (H - OH) / Ca) / (1.0 + (H - OH) / Cb)
        
        # Cálculo del volumen del titulante
        V_b = Fi * (Ca * Va) / Cb
        Vb_valores.append(V_b)
    
    return Vb_valores, pH_valores

# Solicitar al usuario los valores de concentración y volumen del ácido y la base
Ca = float(input("Ingresa la concentración del ácido M []: "))
Va = float(input("Ingresa el volumen del ácido (mL): "))
Cb = float(input("Ingresa la concentración de la base M []: "))
Vb = float(input("Ingresa el volumen de la base (mL): "))

Vb_valores, pH_valores = calculate_titration_curve(Ca, Va, Cb, Vb)

# Gráfico de la curva de titulación
plt.plot(Vb_valores, pH_valores)
plt.title('Curva de titulación')
plt.xlabel('Volumen B (mL)')
plt.ylabel('pH')
plt.grid(True)
plt.show()
```
