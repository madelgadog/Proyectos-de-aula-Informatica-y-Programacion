```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_titration_curve(Ca, Va, Cb, Vb):
    # pKa de 치cido formico
    pKa = 3.75
    Kw = 1.0e-14
    
    # C치lculo de la constante de disociaci칩n 치cida
    Ka = 10.0**(-pKa)
    
    # Rango de pH y n칰mero de puntos
    pH_ini = 0
    pH_fin = 14
    puntos = 100
    pH_valores = np.linspace(pH_ini, pH_fin, puntos)
    
    Vb_valores = []
    for pH in pH_valores:
        H = 10**(-pH)
        OH = Kw / H
        
        # C치lculo de la fracci칩n de titulaci칩n
        alpha_A = Ka / (H + Ka)
        Fi = (alpha_A - (H - OH) / Ca) / (1.0 + (H - OH) / Cb)
        
        # C치lculo del volumen del titulante
        V_b = Fi * (Ca * Va) / Cb
        Vb_valores.append(V_b)
    
    return Vb_valores, pH_valores

# Solicitar al usuario los valores de concentraci칩n y volumen del 치cido y la base
Ca = float(input("Ingresa la concentraci칩n del 치cido M []: "))
Va = float(input("Ingresa el volumen del 치cido (mL): "))
Cb = float(input("Ingresa la concentraci칩n de la base M []: "))
Vb = float(input("Ingresa el volumen de la base (mL): "))

Vb_valores, pH_valores = calculate_titration_curve(Ca, Va, Cb, Vb)

# Gr치fico de la curva de titulaci칩n
plt.plot(Vb_valores, pH_valores)
plt.title('Curva de titulaci칩n')
plt.xlabel('Volumen B (mL)')
plt.ylabel('pH')
plt.grid(True)
plt.show()
```
