```python
from sympy import *
from sympy.plotting import plot
from sympy.abc import x
init_printing()
f1 = cos(x)
df = diff(f1, x) # Primera derivada
d2f = diff(f1,x,2) # Segunda derivada
df, d2f
pc = solve(Eq(df, 0))
pc
d2f.subs(x,pc[0]) # Primer punto crítico
d2f.subs(x,pc[0]) # Segundo punto crítico
plot(f1, (x, -10, 10))
def maxminf(f1):
    """ Calcula los máximos y mínimos de una función f(x) """
    df = diff(f1, x) # 1era. derivada
    d2f = diff(f1, x, 2) # 2da. derivada
    pcs = solve(Eq(df,0)) # puntos críticos
    for p in pcs:
        if d2f.subs(x,p)>0: 
            tipo="Min"
        elif d2f.subs(x,p)<0: 
            tipo="Max"
        else: 
            tipo="Indefinido"
        print("x = %f1 (%s)"%(p,tipo))
maxminf(f1)
```
