```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp

def g(x):
    # Función de punto fijo
    return (2*exp(x**2))/5

def punto_fijo(x0, eps):
    # Método del punto fijo para encontrar la raíz
    iteraciones = []
    errores = []
    while True:
        x1 = g(x0)
        iteraciones.append(x1)
        error = abs(x1 - x0)
        errores.append(error)
        if error < eps:
            plt.plot(iteraciones, '-o')
            plt.xlabel("Iteración")
            plt.ylabel("Valor de x")
            plt.show()
            return x1
        x0 = x1

# Ejemplo de uso
raiz = punto_fijo(0, 1e-2)
print("La raíz es:", raiz)
```
